```python
# graph data
def make_fig(fname,savename,start_time=None,end_time=None):
    curves = Berry.row(fname)
    subset_curves = deepcopy(curves)
    for name in curves:
        if name not in names:
            subset_curves.pop(name)
    for name,curve in subset_curves.items():
        n = curve.normalize()
        if start_time is not None and end_time is not None:
            n = curve.trim(start_time,end_time)
        times,vals = zip(*n.items())
        plt.plot(times,vals,label=name)
    plt.legend()
    plt.ylabel("normalized expression")
    plt.xlabel("times")
    plt.savefig(savename) 
    display.display(plt.show())
    
```

```python
make_fig(wt1_file,"time_series_rep1.pdf")
make_fig(wt2_file,"time_series_rep2.pdf")
```

```python
def make_posets(p,network):
    events = list(p[1][0])
    event_ordering = list(p[1][1]) 
    poe = DSGRN.PosetOfExtrema(network, events, event_ordering)
    return poe,len(events)
```

```python
def create_pattern_graph(poset,network):
    eps = poset[0]
    poe,_ = make_posets(poset,network)
    return DSGRN.PatternGraph(poe),eps
```

```python
def create_search_graph(param):
    domain_graph = DSGRN.DomainGraph(param)
    return DSGRN.SearchGraph(domain_graph)  
```

```python
def get_matches(posets,network):
    param_matches = {}
    parameter_graph = DSGRN.ParameterGraph(network)
    for poset in posets:
        pattern_graph,eps = create_pattern_graph(poset,network)
        param_matches[eps] = []
        for pind in range(parameter_graph.size()):
            parameter = parameter_graph.parameter(pind)
            search_graph = create_search_graph(parameter)
            matching_graph = DSGRN.MatchingGraph(search_graph, pattern_graph)
            path_match = DSGRN.PathMatch(matching_graph)
            if path_match:
                param_matches[eps].append(pind)
    return param_matches
```

```python
wavepool1_parameter_matches=get_matches(posets1,wavepool)
wavepool2_parameter_matches=get_matches(posets2,wavepool)
swapped11_parameter_matches=get_matches(posets1,swapped)
swapped12_parameter_matches=get_matches(posets2,swapped)

for eps in epsilons:
    print("Epsilon = {}".format(eps))
    print("Number of matches in rep 1, wavepool: {}".format(len(wavepool1_parameter_matches[eps])))
    print("Number of matches in rep 2, wavepool: {}".format(len(wavepool2_parameter_matches[eps])))
    print("Number of matches in rep 1, swapped: {}".format(len(swapped11_parameter_matches[eps])))
    print("Number of matches in rep 2, swapped: {}".format(len(swapped12_parameter_matches[eps])))
```

## Application 2

```python
start_time = 0
end_time = 170

make_fig(wt1_file,"time_series_rep1_trimmed.pdf",start_time,end_time)
make_fig(wt2_file,"time_series_rep2_trimmed.pdf",start_time,end_time)
```

```python
epsilons = [0.005*x for x in range(32)]

posets1 = Berry.getposets(wt1_file,"row",epsilons,names,start_time,end_time)
posets2 = Berry.getposets(wt2_file,"row",epsilons,names,start_time,end_time)
posets3 = Berry.getposets(wt2_file,"row",epsilons,["CLB2","SWI4","HCM1","NDD1"],start_time,end_time)
```

```python
def change_labels(posets,start_name,end_name):
    relabeled_posets = []
    for eps,pos in posets:
        extrema = pos[0]
        newex = []
        for ex in extrema:
            if ex[0] == start_name:
                newex.append((end_name,ex[1]))
            else:
                newex.append(ex)
        newpos = (tuple(newex),pos[1])
        relabeled_posets.append((eps,newpos))
    return relabeled_posets
                  
```

```python
posets4 = change_labels(posets3,"CLB2","YOX1")
```

```python
def calc_dists(posets1,posets2):
    dists = []
    sizes = []
    for p1,p2 in zip(posets1,posets2):
        g1 = poset_distance.poset_to_nx_graph(p1[1])
        g2 = poset_distance.poset_to_nx_graph(p2[1])
        sizes.append((len(g1.nodes())+len(g1.edges()), len(g2.nodes())+len(g2.edges())))
        dists.append(poset_distance.dag_distance(g1,g2))
    return dists,sizes
```

```python
def plot_dists(epsilons,dists,label,ylabel="Poset distance"):
    plt.plot(epsilons,dists,linewidth=2,label=label)
    plt.xlabel(r"$\mathrm{\epsilon}$",fontsize=16)
    plt.ylabel(ylabel)
    plt.legend()
```

```python
dists12,sizes12 = calc_dists(posets1,posets2)
dists13,sizes13 = calc_dists(posets1,posets3)
dists14,sizes14 = calc_dists(posets1,posets4)
```

```python
sumsizes12 = (a+b for a,b in sizes12)
sumsizes13 = (a+b for a,b in sizes13)
sumsizes14 = (a+b for a,b in sizes14)
normdists12 = [d/s for d,s in zip(dists12,sumsizes12)]
normdists13 = [d/s for d,s in zip(dists13,sumsizes13)]
normdists14 = [d/s for d,s in zip(dists14,sumsizes14)]
```

```python
print("Distance with same genes.")
plot_dists(epsilons,dists12,ylabel="Poset distance",label="YOX1/YOX1")
print("Distance with one swapped gene.")
plot_dists(epsilons,dists13,ylabel="Poset distance",label="YOX1/CLB2")
print("Distance with one swapped gene relabeled.")
plot_dists(epsilons,dists14,ylabel="Poset distance",label="CLB2 relabeled")
plt.savefig("poset_distances.pdf") 
```

```python
print("Distance with same genes.")
plot_dists(epsilons,normdists12,ylabel="Normalized poset distance",label="YOX1/YOX1")
print("Distance with one swapped gene.")
plot_dists(epsilons,normdists13,ylabel="Normalized poset distance",label="YOX1/CLB2")
print("Distance with one swapped gene relabeled.")
plot_dists(epsilons,normdists14,ylabel="Normalized poset distance",label="CLB2 relabeled")
plt.savefig("poset_distances_normalized.pdf") 
```
